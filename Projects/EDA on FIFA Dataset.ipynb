### Print the highest salary of players in euro

```python
player_salary = fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

### Sort the salary according to the descending order

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head()
```

### Graphically representation of player and its salary in euro

```python
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=["blue","green","red","pink","orange"])
plt.show()
```
