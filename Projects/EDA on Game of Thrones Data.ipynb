# EDA on Game of Thrones Data

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

### Added data set of battles

```python
battle = pd.read_csv(r'C:\\Users\munee\Downloads\Data-Science-Course-main\battles.csv')
```

```python
battle.head()
```

```python
battle.shape
```

### Replace all NaN column to None

```python
battle.isnull().sum()
```

```python
battle.replace(np.nan, "None",inplace=True)
```

```python
battle.isnull().sum()
```

```python
battle.head()
```

### Rename attacker 1 as primary attracker

```python
battle.rename(columns={'attacker_1':'primary_attacker'},inplace=True)
battle.head()
```

### Rename defender 1 as primary defender

```python
battle.rename(columns={'defender_1':'primary_defender'},inplace=True)
battle.head()
```

### count the value of attacker king

```python
battle['attacker_king'].value_counts()
```

### count the value of locations

```python
battle['location'].value_counts()
```

### Graphically representation of attacker king and attacker size

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='attacker_king',y='attacker_size',data=battle)
plt.show()
```

### Graphically representation of defender king and defender size

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='defender_king',y='defender_size',data=battle)
plt.show()
```

### Graphically representation of battle types

```python
sns.countplot(x=battle['attacker_king'],hue=battle['battle_type'])
plt.show()
```
